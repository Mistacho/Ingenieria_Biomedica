**Fase 4 Creación: El Desafío de Diseño (15 min)**

Diseña una función llamada analizar_signos_vitales que cumpla estos requisitos:
1. Recibe dos parámetros: ritmo_cardiaco (obligatorio) y limite_alto (opcional, valor por defecto = 100).
2. Si el ritmo supera el límite, devuelve True (Alerta). Si no, devuelve False.
3. Debe incluir un Docstring.


```python
def analizar_signos_vitales(ritmo_cardiacio, limite_alto=100):
    """
        Esta función compara el ritmo_cardiaco de un paciente y establece una alerta si supera el limite establecido.
        Parametros: 
            ritmo_cardiaco: ritmo caridaco del paciente (obligatorio)
            limite_alto: límite a partir del cual suena una alerta (por defecto: 100)
        Return: Devuelve True o False dependiendo de si el ritmo supera el límite o no
    """
    alerta = False
    if ritmo_cardiacio > limite_alto:
        alerta = True 
    return alerta
```
