Escribe un programa nuevo desde cero llamado "Control de Acceso al Laboratorio". Debe cumplir las siguientes reglas:
1. Pedir al usuario su cargo (Estudiante, Profesor, Investigador).
2. Pedir al usuario la hora actual (entero de 0 a 23).
3. Reglas de acceso:
    ◦ Los "Investigadores" pueden entrar a cualquier hora.
    ◦ Los "Profesores" pueden entrar solo entre las 7 y las 22 horas.
    ◦ Los "Estudiantes" solo pueden entrar si es entre las 8 y las 20 horas y van acompañados (preguntar: "¿Vienes acompañado? Sí/No").
4. Imprimir "Acceso Permitido" o "Acceso Denegado".

```python
# Código Base: Control de Acceso al Laboratorio
print("--- Sistema de  Control de Acceso al Laboratorio ---")

cargo = input("Introduzca su cargo (Estudiante, Profesor, Investigador)")
horaActual = int(input("Introduzca la hora actual (0 a 23)"))


if(cargo == "Profesor" or cargo == "Estudiante" or cargo == "Investigador"):
    if (cargo == "Profesor" and (horaActual < 7 or horaActual > 22)):
        print("Acceso Denegado")
    elif(cargo == "Estudiante"):
        acompañado = input("Vienes acompañado? (Si/No)")        
        if(acompañado == "No" or acompañado == "Si"):
            if(acompañado == "No" or (acompañado == "Si" and (horaActual < 8 or horaActual > 20))):
               print("Acceso Denegado") 
            else:
                print("Acceso Permitido")
        else: 
            print("Error, responda Si o No a la pregunta de si viene acompañado")
    else:
        print("Acceso Permitido")
else:
    print("Error, cargo no encontrado")
```
