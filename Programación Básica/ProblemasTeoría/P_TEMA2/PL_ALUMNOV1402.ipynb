# NOTA: EFICIENCIA DE WHILE VS EFICIENCIA DE FOR

```python
def sumaWhile(num):
    total=0
    count=1
    while count < num:
        total += count
        count += 2
    return(total)

def sumaFor(num):
    total=0
    for i in range(1, num,2):
        total+=i
    return(total)

%timeit -n 1000 sumaWhile(100)
%timeit -n 1000 sumaFor(100)
```
