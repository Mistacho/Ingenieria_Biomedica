# Ejemplo de teoría: dos algoritmos y cuatro programas para calcular n^exp

```python
# algoritmo 1 con while

def algoritmo1while(n,exp):
    cont=1
    resultado=1
    while cont<=exp:
        resultado=resultado*n
        cont+=1
    return(resultado)
```

```python
# mismo algoritmo 1 con for
def algoritmo1for(n,exp):
    resultado=1
    for cont in range(1,exp+1):
        resultado=resultado*n
    return(resultado)
```

```python
# algoritmo 2 con while
def algoritmo2while(n,exp):
    cont=1
    resultado=n
    it=int(pow(exp,0.5))
    #print(it)
    while cont<=it:
        resultado=resultado*resultado
        cont+=1
    resto=exp-pow(2,it)
    #print(resto)
    i=0
    while i<resto:
        resultado=resultado*n
        i=i+1
    return(resultado)
```

```python
# mismo algoritmo 2 con for
def algoritmo2for(n,exp):
    resultado=n
    it=int(pow(exp,0.5))
    for i in range(0,it):
        resultado=resultado*resultado
    resto=exp-pow(2,it)
    for i in range(0,resto):
        resultado=resultado*n
    return(resultado)



```

```python
n=int(input("Dime num: "))
exp=int(input("Dime exponente: "))
%timeit algoritmo1while(n,exp)
%timeit algoritmo1for(n,exp)
%timeit algoritmo2while(n,exp)
%timeit algoritmo2for(n,exp)
print("Resultado=",algoritmo2for(n,exp))
```
