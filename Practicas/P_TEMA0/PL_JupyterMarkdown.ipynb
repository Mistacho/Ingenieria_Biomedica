# Lenguaje de Marcado de textos (Markup language)


## Títulos

En los jupyter notebooks se pueden introducir porciones de textos con su lenguaje de markdown:

# Título
## Subtítulo
### Subsubtítulo
#### Subsubsubtítulo
##### ...



## Texto en cursiva y negrita

Tb puedo escribir texto en _cursiva_ y en __negrita__

The Markdown parser included in the Jupyter Notebook is MathJax-aware. This means that you can freely mix in mathematical expressions using the MathJax subset of Tex and LaTeX. Some examples from the MathJax site are reproduced below, as well as the Markdown+TeX source. (Fuente: http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Typesetting%20Equations.html)

\begin{equation*}
\left( \sum_{k=1}^n a_k b_k \right)^2 \leq \left( \sum_{k=1}^n a_k^2 \right) \left( \sum_{k=1}^n b_k^2 \right)
\end{equation*}

## ¿Cómo escribimos listas?
### Listas no numeradas: *


* Elemento 1
* Elemento 2
    * Subelemento 1
    * Subelemento 2
    
### Listas numeradas: con un número.
#### ¡Ojo! ¡No es necesario que sean consecutivos!
1. Elemento numerado 1
3. Elemento numerado 2
    1. Subelemento 1
    *  Subelemento3
