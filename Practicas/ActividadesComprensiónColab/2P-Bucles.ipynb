A continuación se presenta un algoritmo que analiza una lista de voltajes (en milivoltios) capturados por un sensor cardíaco. El objetivo es detectar "picos R" (latidos) que superan un umbral. NO EJECUTES EL CÓDIGO AÚN. Lee el código y completa la "Tabla de Trazado" en la celda de texto habilitada al efecto o en papel.

```python
# Señal de ECG simulada (mV)
# Representa 10 lecturas consecutivas del sensor
lecturas_ecg = [0.2, 0.5, 1.2, 0.4, 0.1, 0.3, 1.5, 0.2, 0.0, 1.1]

umbral_latido = 1.0
contador_latidos = 0
suma_voltajes = 0

print("Iniciando análisis...")

for lectura in lecturas_ecg:
    suma_voltajes = suma_voltajes + lectura # Acumulador

    if lectura > umbral_latido:
        contador_latidos += 1 # Contador
        print(f"Latido detectado: {lectura} mV")

promedio = suma_voltajes / len(lecturas_ecg)
print(f"Total latidos: {contador_latidos}")
print(f"Voltaje promedio: {promedio}")
```
