```python
# Código Base: Sistema de Triaje Simplificado
print("--- Sistema de Triaje de Emergencias ---")

temp_input = input("Ingrese temperatura del paciente (°C): ")
temperatura = float(temp_input)

sat_input = input("Ingrese saturación de oxígeno (%): ")
saturacion = int(sat_input)

nivel_urgencia = "Bajo"

if temperatura > 38.0 or saturacion < 90:
    nivel_urgencia = "Alto"
    print("¡ALERTA CLÍNICA!")
elif temperatura > 37.5:
    nivel_urgencia = "Medio"
else:
    nivel_urgencia = "Bajo"

print("Resultado del Triaje: Prioridad " + nivel_urgencia)
```
