**Fase 3: Modificación - "Scaffolding" (20 min)**

El laboratorio te pide calcular el Porcentaje de Contenido GC (Guanina + Citosina) de una secuencia. Es un indicador de estabilidad térmica del ADN. Tienes el esqueleto del código, rellena los huecos (___).

```python
adn = "agcttttcattctgactgcaacgkkk" # Nota: hay 'k' (ruido) y está en minúsculas

# Paso 1: Normalizar a mayúsculas
adn_upper = adn.___() # Usa el método para mayúsculas [20]

# Paso 2: Limpiar caracteres inválidos (solo queremos A, C, G, T)
# PISTA: Las 'k' son errores de lectura. Reemplázalas por vacío.
adn_limpio = adn_upper.replace("___", "")

# Paso 3: Contar bases
num_c = adn_limpio.count("___")
num_g = adn_limpio.___("G") # Completa el método

# Paso 4: Calcular longitud total
total_bases = ___(adn_limpio) # Función built-in para longitud [21]

# Paso 5: Calcular porcentaje
porcentaje_gc = (num_c + num_g) / total_bases * 100

print(f"Contenido GC: {porcentaje_gc:.2f}%")
```
