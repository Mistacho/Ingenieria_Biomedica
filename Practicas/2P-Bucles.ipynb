**Fase 1: PREDICT (Predecir) - 10 Minutos**
A continuación se presenta un algoritmo que analiza una lista de voltajes (en milivoltios) capturados por un sensor cardíaco. El objetivo es detectar "picos R" (latidos) que superan un umbral. NO EJECUTES EL CÓDIGO AÚN. Lee el código y completa la "Tabla de Trazado" en la celda de texto habilitada al efecto o en papel.

```python
# Señal de ECG simulada (mV)
# Representa 10 lecturas consecutivas del sensor
lecturas_ecg = [0.2, 0.5, 1.2, 0.4, 0.1, 0.3, 1.5, 0.2, 0.0, 1.1]

umbral_latido = 1.0
contador_latidos = 0
suma_voltajes = 0

print("Iniciando análisis...")

for lectura in lecturas_ecg:
    suma_voltajes = suma_voltajes + lectura # Acumulador

    if lectura > umbral_latido:
        contador_latidos += 1 # Contador
        print(f"Latido detectado: {lectura} mV")

promedio = suma_voltajes / len(lecturas_ecg)
print(f"Total latidos: {contador_latidos}")
print(f"Voltaje promedio: {promedio}")
```

**Fase 3: Modificación y Control de Flujo (15 Minutos)**
Los sensores reales fallan. A veces envían ruido (valores negativos imposibles en este contexto simplificado) o se desconectan (envían ceros consecutivos).
Modifica el código anterior en una nueva celda para cumplir dos nuevas reglas de seguridad:
1. Filtrado de ruido (continue): Si la lectura es negativa (< 0), ignórala y pasa a la siguiente medición. No debe sumar al promedio ni contar como latido.
2. Parada de emergencia (break): Si la lectura es exactamente 99.9 (código de error de sensor desconectado), el bucle debe detenerse inmediatamente e imprimir "ALERTA: Sensor desconectado".

Datos de prueba sugeridos: [0.2, -0.5, 1.2, 99.9, 0.3]

```python
# Señal de ECG simulada (mV)
# Representa 10 lecturas consecutivas del sensor
lecturas_ecg = [0.2, -0.5, 1.2, 99.9, 0.3]

umbral_latido = 1.0
contador_latidos = 0
suma_voltajes = 0

print("Iniciando análisis...")

for lectura in lecturas_ecg:
    if(lectura < 0):
        continue
    elif(lectura == 99.9):
        print("ALERTA: Sensor desconectado")
        break

    suma_voltajes = suma_voltajes + lectura # Acumulador

    if lectura > umbral_latido:
        contador_latidos += 1 # Contador
        print(f"Latido detectado: {lectura} mV")


promedio = suma_voltajes / len(lecturas_ecg)
print(f"Total latidos: {contador_latidos}")
print(f"Voltaje promedio: {promedio}")
```

**Fase 4: Conversión de Estructuras (15 Minutos)**

Reescribe el analizador de ECG original (Fase 1) utilizando un bucle while en lugar de un for.

Pistas:
- Necesitarás gestionar manualmente una variable índice (ej. i = 0).
- La condición del while debe basarse en la longitud de la lista (len(lecturas_ecg)).
- ¡No olvides incrementar el índice (i += 1) al final del bucle o tendrás un bucle infinito!.

```python
# Señal de ECG simulada (mV)
# Representa 10 lecturas consecutivas del sensor
lecturas_ecg = [0.2, 0.5, 1.2, 0.4, 0.1, 0.3, 1.5, 0.2, 0.0, 1.1]

umbral_latido = 1.0
contador_latidos = 0
suma_voltajes = 0

print("Iniciando análisis...")

#Creamos un contador para poder salir del bucle while
i = 0
while (i < len(lecturas_ecg)):
    lectura = lecturas_ecg[i]
    suma_voltajes = suma_voltajes + lectura # Acumulador

    if lectura > umbral_latido:
        contador_latidos += 1 # Contador
        print(f"Latido detectado: {lectura} mV")
    
    i+= 1

promedio = suma_voltajes / len(lecturas_ecg)
print(f"Total latidos: {contador_latidos}")
print(f"Voltaje promedio: {promedio}")
```

Basándote en esto: ¿Por qué preferimos usar for para analizar datos históricos (como una lista de latidos ya grabada) y reservamos el while para monitorización en tiempo real?"

**Solucion ejercicio 5**

Para el ejercicio anterior hemos usado un for porque sabiamos perfectamente el número de iteraciones que ibamos a realizar, ya que coincide con la longitud de la lista que tenemos pasada por parametro (lecturas_ecg). Sin embargo, para monitorizacion a tiempo real como no es una lista "finita", sino una en la que se van añadiendo valores constantemente no sabemos cuando se parará (cuantas iteraciones totales haremos) y usamos un while en su defecto.
