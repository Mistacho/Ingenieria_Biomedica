# 0: CÁLCULO NOTA EVALUACIÓN CONTINUA

Sobre cinco puntos: 
*   0.75 puntos cada test (total: 1.5 puntos)
*   0.25 puntos cada entrega en el servidor (total: 0.5 puntos)
*   1 punto el examen del tema 2
*   2 puntos el examen del tema 4 (nota examen + 0.5 test IA)

Sobre 10, la puntuación sería el doble para cada apartado.



```python

notaTest3=float(input("Introduce la nota del test del tema 3 (0..10): "))
notaTest5=float(input("Introduce la nota del test del tema 5 (0..10): "))

notaEntrega5=float(input("Introduce la nota de la entrega del servidor de prácticas del tema 5 (estructuras de datos) (0..10): "))
notaEntrega6=float(input("Introduce la nota de la entrega del servidor de prácticas del tema 6 (ficheros) (0..10): "))

notaSobre2=(0.75*notaTest3+0.75*notaTest5+0.25*notaEntrega5+0.25*notaEntrega6)/10
print(notaSobre2)

notaExamenTema2=float(input("Introduce la nota de la corrección colaborativa del tema 2 (bucles) (0..5): "))
notaCorreccionTema2=float(input("Introduce la nota de tu labor como corrector en el tema 2 (bucles) (0..5): "))
notaTotalExamenTema2=(notaExamenTema2+notaCorreccionTema2)/10
print(notaTotalExamenTema2)

notaExamenTema4=float(input("Introduce la nota de la corrección colaborativa del tema 4 (funciones) (0..6): "))
notaCorreccionTema4=float(input("Introduce la nota de tu labor como corrector en el tema 4 (funciones) (0..4): "))
testIA=int(input("Realizaste el test de IA? (0:NO, 1:SÍ): "))
notaIA=0
if(testIA==1):
  notaIA=0.5
notaTotalExamenTema4=(notaExamenTema4+notaCorreccionTema4+notaIA)*2/10
print(notaTotalExamenTema4)

evalCont=notaSobre2+notaTotalExamenTema2+notaTotalExamenTema4

print("Tu nota sobre 10 en la evaluación continua es: ",round(evalCont,2)*2)
if (round(evalCont,2)>=3.5):
  print("Enhorabuena, tienes que realizar solo el proyecto")
else:
  
  print("Lo siento, debes presentarte el examen. Recuerda que, aún así, puedes realizar el proyecto también para subir nota")
```
